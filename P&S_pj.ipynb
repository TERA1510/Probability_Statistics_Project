### 상관계수의 유의성 검정 방법 (모두 공통)

- 상관계수로는 피어슨 상관 계수를 사용한다. 상관계수 $\rho$를 구하는 수식은 다음과 같다.

<br>

$$\rho = \frac{Cov(X, Y)}{\sigma(X)\sigma(Y)}$$

- 검정 과정은 t-검정을 사용한다. $n$은 표본 개수, $\rho$는 표본의 상관계수, $\rho_0$는 귀무가설에서 가정하는 모상관계수이다.

<br>

$$
\frac{\rho - \rho_0}{\sqrt{\frac{1-\rho^2}{n-2}}} \sim t(n-2)
$$

<br>

__(1) 가설 설정__
- 귀무가설 ($H_0$): $X$, $Y$ 두 변수 간의 상관계수가 0이다 ($\rho=0$). 즉, 선형 관계가 없다.
- 대립가설 ($H_1$): $X$, $Y$ 두 변수 간의 상관계수가 0이 아니다($\rho \neq 0$). 즉, 선형 관계가 있다.

__(2) 검정 통계량 계산__
- t-검정 통계량을 계산하기 위해 다음 수식을 사용한다.
- $\rho_0$은 0으로 둔다.


$$
t = \frac{\rho\sqrt{n-2}}{\sqrt{1-\rho^2}}
$$

__(3) p-value 계산__
- p-value는 제 1종 오류를 범하게 될 확률로, 실제로는 두 변수 간의 상관계수가 0인데 표본의 상관계수가 있는 것으로 드러날 확률이다.
- t-분포와 계산된 t-값에 대한 양측 검정 p-value를 계산한다. p-value는 다음과 같이 계산된다.

$$
p = 2 \cdot P(T > |t|)
$$

- 이때, $P(T>|t|)$ 는 자유도가 $n-2$인 t-분포 $t(n-2)$ 에서 $|t|$ 이상의 값이 나올 확률이다.
- t-분포의 누적분포함수(CDF)를 사용하여 p-value를 계산한다.

$$
p = 2 \cdot (1 - t.cdf(|t|, n-2))
$$


__(4) 유의수준 설정 및 p-value 확인__
- 99% 신뢰도로 검정을 진행하기 위해, 유의수준 $\alpha = 0.01$로 설정한다.
- p-value가 $\alpha$보다 작으면, 귀무가설이 기각되어 선형 관계가 있다고 할 수 있다.

- 여러 전처리 과정을 통해 서로 다른 선형회귀 결과가 나올 경우, p_value가 더 작고, 상관계수가 더 증가할 때 선형회귀의 유의성이 더 높다고 할 수 있다.

### 소장 도서 수와 대출권수 간의 상관관계 분석 (+검정)

- X(독립변수)는 소장 도서 수, Y(종속변수)는 대출 권수.
- numpy를 활용하여 직접 피어슨 상관계수를 계산 후 출력 (소숫점 아래 20자리까지)
- 선형회귀를 진행한 후 기울기와 y절편 출력 (소숫점 아래 20자리까지)
- X, Y로 scatter plot을 그리고, 선형회귀 결과 얻은 직선을 scatter plot 위에 그림.
- 서울과학고등학교의 x, y 데이터를 구한 후, 이 데이터를 scatter plot 위에 강조

우선, 데이터에 특별한 전처리를 진행하지 않고 바로 상관관계 분석을 해보자.

```python
# 데이터 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.parse
import koreanize_matplotlib

raw_url = 'https://github.com/TERA1510/Probability_Statistics_Project/raw/refs/heads/master/문화체육관광부_국가도서관통계_전국학교도서관통계_20231231.csv'
url = urllib.parse.quote(raw_url, safe=':/')
school_library_stats = pd.read_csv(url, encoding='euc-kr')
school_library_stats.head()
```

```python
# 데이터 전처리. 대출 권수가 0인 행 제거
school_library_stats = school_library_stats[school_library_stats['대출권수'] > 0]
```

```python
from sklearn.linear_model import LinearRegression

X = school_library_stats['장서수(인쇄)']
Y = school_library_stats['대출권수']

# 피어슨 상관계수 계산
pearson_corr = np.corrcoef(X, Y)[0, 1]
print(f"Pearson Correlation Coefficient: {pearson_corr:.20f}")

# 검정 통계량 t 계산
t_stat = pearson_corr * np.sqrt((len(X) - 2) / (1 - pearson_corr**2))
print(f"t-statistic: {t_stat:.20f}")

# p-value 계산
from scipy.stats import t
p_value = 2 * (1 - t.cdf(abs(t_stat), df=len(X) - 2))
print(f"p-value: {p_value:.20f}")

# 유의수준 0.01로 검정 수행
if p_value < 0.01:
    print("귀무가설을 기각하고 대립가설을 채택합니다. 선형 관계가 있습니다.")
else:
    print("귀무가설을 채택합니다. 선형 관계가 없습니다.")
print()

# Perform linear regression
X_reshaped = X.values.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, Y)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.20f}")
print(f"Intercept: {intercept:.20f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6), dpi=200)
plt.scatter(X, Y, label='데이터 포인트', alpha=0.7, color='orange', s=1)
plt.plot(X, model.predict(X_reshaped), color='red', label='회귀선', lw=1)
plt.xlabel('소장 도서 수')
plt.ylabel('대출 권수')
plt.title('소장 도서 수 vs 대출 권수')

# 서울과학고등학교 데이터 강조
sshs_data = school_library_stats[school_library_stats['도서관명'] == '서울과학고등학교']
sshs_data_x = sshs_data['장서수(인쇄)'].values[0]
sshs_data_y = sshs_data['대출권수'].values[0]
print(f"서울과학고등학교의 장서수(인쇄): {sshs_data_x}, 대출권수: {sshs_data_y}")
plt.scatter([sshs_data_x], [sshs_data_y], color='green', s=60, label='서울과학고등학교', marker='x')
plt.annotate('서울과학고등학교', (sshs_data_x, sshs_data_y), textcoords="offset points", xytext=(-10, 10), ha='center')

plt.grid(True)
plt.legend()
plt.show()
```

- 데이터 포인트를 찍어보니, 데이터의 추세가 log 형태임을 알게 되었다.
- 좀 더 나은 상관계수를 얻어보기 위해, 데이터에 log를 씌우고 분석을 추가로 진행하였다.

```python
from sklearn.linear_model import LinearRegression

X = school_library_stats['장서수(인쇄)']
Y = np.log(school_library_stats['대출권수']) # log(0)을 방지하기 위해 1을 더함

# 피어슨 상관계수 계산
pearson_corr = np.corrcoef(X, Y)[0, 1]
print(f"Pearson Correlation Coefficient: {pearson_corr:.20f}")

# 검정 통계량 t 계산
t_stat = pearson_corr * np.sqrt((len(X) - 2) / (1 - pearson_corr**2))
print(f"t-statistic: {t_stat:.20f}")

# p-value 계산
from scipy.stats import t
p_value = 2 * (1 - t.cdf(abs(t_stat), df=len(X) - 2))
print(f"p-value: {p_value:.20f}")

# 유의수준 0.01로 검정 수행
if p_value < 0.01:
    print("귀무가설을 기각하고 대립가설을 채택합니다. 선형 관계가 있습니다.")
else:
    print("귀무가설을 채택합니다. 선형 관계가 없습니다.")
print()

# Perform linear regression
X_reshaped = X.values.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, Y)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.20f}")
print(f"Intercept: {intercept:.20f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6), dpi=200)
plt.scatter(X, Y, label='데이터 포인트', alpha=0.7, color='orange', s=1)
plt.plot(X, model.predict(X_reshaped), color='red', label='회귀선', lw=1)
plt.xlabel('소장 도서 수')
plt.ylabel('log(대출 권수)')
plt.title('소장 도서 수 vs log(대출 권수)')

# 서울과학고등학교 데이터 강조
sshs_data = school_library_stats[school_library_stats['도서관명'] == '서울과학고등학교']
sshs_data_x = sshs_data['장서수(인쇄)'].values[0]
sshs_data_y = np.log(sshs_data['대출권수'].values[0])
print(f"서울과학고등학교의 장서수(인쇄): {sshs_data_x}, log(대출권수): {sshs_data_y}")
plt.scatter([sshs_data_x], [sshs_data_y], color='green', s=60, label='서울과학고등학교', marker='x')
plt.annotate('서울과학고등학교', (sshs_data_x, sshs_data_y), textcoords="offset points", xytext=(-10, 10), ha='center')

plt.grid(True)
plt.legend()
plt.show()
```

- 대출 권수에 log를 씌우지 않았을 때와 씌웠을 때 모두 p-value가 0.000000 으로 통계적으로 매우 유의미
- 단, log를 씌우지 않았을 때는 상관 계수가 0.479
- log를 씌웠을 때는 상관 계수가 0.567
- 둘 모두 상관관계가 있다!

```python
# 데이터 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.parse
import koreanize_matplotlib

raw_url = 'https://github.com/TERA1510/Probability_Statistics_Project/raw/refs/heads/master/교육부_초중고 학교도서관 현황_학교별_2022_2024.csv'
url = urllib.parse.quote(raw_url, safe=':/')
school_library_stats2 = pd.read_csv(url, encoding='utf8')
school_library_stats2.info()
```

```python
# 데이터 전처리
# 연도 2023년인 행만 선택
school_library_stats2 = school_library_stats2[school_library_stats2['연도'] == 2023]

# merge 위해 standarize
school_library_stats['도서관명'] = school_library_stats['도서관명'].str.strip()
school_library_stats2['학교명'] = school_library_stats2['학교명'].str.strip()

# Perform a left merge to add 예산액_계 from school_library_stats2 to school_library_stats
merged_data = pd.merge(
    school_library_stats,
    school_library_stats2[['학교명', '행정구', '예산액_계']],
    left_on=['도서관명', '시군구'],
    right_on=['학교명', '행정구'],
    how='inner'
)

# Drop the duplicate '학교명' column
merged_data = merged_data.drop(columns=['학교명'])
merged_data = merged_data.rename(columns={'예산액_계': '예산액'})
merged_data['예산액'] = merged_data['예산액'].str.replace(',', '').astype(float)
merged_data = merged_data.dropna(subset=['예산액', '대출자수', '대출권수'])
merged_data['예산액'] = merged_data['예산액'].astype(int)
merged_data = merged_data[merged_data['대출권수'] > 0]
merged_data = merged_data[merged_data['대출자수'] > 0]

merged_data.info()
```

```python
from sklearn.linear_model import LinearRegression

X = merged_data['예산액']
Y = merged_data['대출권수']

# 피어슨 상관계수 계산
pearson_corr = np.corrcoef(X, Y)[0, 1]
print(f"Pearson Correlation Coefficient: {pearson_corr:.20f}")

# 검정 통계량 t 계산
t_stat = pearson_corr * np.sqrt((len(X) - 2) / (1 - pearson_corr**2))
print(f"t-statistic: {t_stat:.20f}")

# p-value 계산
from scipy.stats import t
p_value = 2 * (1 - t.cdf(abs(t_stat), df=len(X) - 2))
print(f"p-value: {p_value:.20f}")

# 유의수준 0.01로 검정 수행
if p_value < 0.01:
    print("귀무가설을 기각하고 대립가설을 채택합니다. 선형 관계가 있습니다.")
else:
    print("귀무가설을 채택합니다. 선형 관계가 없습니다.")
print()

# Perform linear regression
X_reshaped = X.values.reshape(-1, 1)
model = LinearRegression()
model.fit(X_reshaped, Y)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.20f}")
print(f"Intercept: {intercept:.20f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6), dpi=200)
plt.scatter(X, Y, label='데이터 포인트', alpha=0.7, color='orange', s=1)
plt.plot(X, model.predict(X_reshaped), color='red', label='회귀선', lw=1)
plt.xlabel('예산액')
plt.ylabel('대출 권수')
plt.title('예산액 vs 대출 권수')

# 서울과학고등학교 데이터 강조
sshs_data = merged_data[merged_data['도서관명'] == '서울과학고등학교']
sshs_data_x = sshs_data['예산액'].values[0]
sshs_data_y = sshs_data['대출권수'].values[0]
print(f"서울과학고등학교의 예산액: {sshs_data_x}, 대출권수: {sshs_data_y}")
plt.scatter([sshs_data_x], [sshs_data_y], color='green', s=60, label='서울과학고등학교', marker='x')
plt.annotate('서울과학고등학교', (sshs_data_x, sshs_data_y), textcoords="offset points", xytext=(-10, 10), ha='center')

plt.grid(True)
plt.legend()
plt.show()
```

- 즉 전체 학교에 비해, 고등학교만 사용해 분석을 하면 약한 상관관계를 가지게 됨을 의미한다.

##학교별 격차 분석

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 로드
url1 = "https://github.com/TERA1510/Probability_Statistics_Project/raw/refs/heads/master/문화체육관광부_국가도서관통계_전국학교도서관통계_20231231.csv"
url2 = "https://github.com/TERA1510/Probability_Statistics_Project/raw/refs/heads/master/교육부_초중고 학교도서관 현황_학교별_2022_2024.csv"
url = urllib.parse.quote(url1, safe=':/')
school_library_stats = pd.read_csv(url, encoding='euc-kr')
# 학교 도서관 통계 데이터 읽기
library_data = school_library_stats

# 학교 기본 정보 데이터 읽기
url = urllib.parse.quote(url2, safe=':/')
school_data = pd.read_csv(url, encoding='utf-8')
school_data.head()
```

```python


# 2. 데이터 정리 및 병합
# 필요한 열 추출 (예: 학교 이름, 지역, 도서관 책 수, 예산 등)
library_data_filtered = library_data[['학교명', '지역', '도서관 책 수', '도서관 예산']]
school_data_filtered = school_data[['학교명', '지역', '학교급']]

# 데이터 병합
merged_data = pd.merge(library_data_filtered, school_data_filtered, on=['학교명', '지역'])

# 3. 격차 분석
# 지역별 도서관 책 수 평균 계산
region_mean_books = merged_data.groupby('지역')['도서관_책_수'].mean()

# 학교급별 도서관 예산 평균 계산
school_type_mean_budget = merged_data.groupby('학교급')['도서관_예산'].mean()

# 4. 시각화
# 지역별 도서관 책 수 시각화
region_mean_books.plot(kind='bar', title='지역별 도서관 책 수 평균', ylabel='책 수', xlabel='지역', figsize=(10, 6))
plt.show()

# 학교급별 도서관 예산 시각화
school_type_mean_budget.plot(kind='bar', title='학교급별 도서관 예산 평균', ylabel='예산 (원)', xlabel='학교급', color='orange', figsize=(10, 6))
plt.show()

# 격차 확인
merged_data['도서관_책_수_편차'] = merged_data['도서관_책_수'] - region_mean_books.loc[merged_data['지역']].values
print(merged_data[['학교명', '지역', '도서관_책_수', '도서관_책_수_편차']].head())
```
